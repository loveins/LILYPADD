```python
import os.path
from pdf2image import convert_from_path

folder = r"C:\Users\valer\OneDrive\Desktop\LIT Hackathon\Emails Invoices and Bank Statements for Hackathon Problem Statement\Bank Statements"

def get_txt_files(folder):
    import os
    os.chdir(folder)
    files = os.listdir()
    files = [x for x in files if x.endswith(".txt")]
    return files 

files = get_txt_files(folder)

for indiv_file in files:
    # file to check
        sz = os.path.getsize(indiv_file)
        
        if sz <= 1:
            filename = indiv_file.split('.')[0]  
            error_file = str(filename) + '.pdf'
            print(error_file)
            
        images = convert_from_path(error_file, 500,poppler_path=r'C:\Users\valer\Downloads\poppler-0.68.0_x86.7z\poppler-0.68.0\bin')
        for i, image in enumerate(images):
            fname = 'image'+str(i)+'.png'
            image.save(fname, "PNG")
```

```python
## convert .doc to .docx ##

from glob import glob
import re
import os
import win32com.client as win32
from win32com.client import constants

# Create list of paths to .doc files
paths = glob(r'C:\Users\valer\OneDrive\Desktop\LIT Hackathon\Emails Invoices and Bank Statements for Hackathon Problem Statement\Emails**\\*.doc', recursive=True)

def save_as_docx(path):
    # Opening MS Word
    word = win32.gencache.EnsureDispatch('Word.Application')
    doc = word.Documents.Open(path)
    doc.Activate ()

    # Rename path with .docx
    new_file_abs = os.path.abspath(path)
    new_file_abs = re.sub(r'\.\w+$', '.docx', new_file_abs)

    # Save and Close
    word.ActiveDocument.SaveAs(
        new_file_abs, FileFormat=constants.wdFormatXMLDocument
    )
    doc.Close(False)

for path in paths:
    save_as_docx(path)

    
## convert .docx to .txt    
    
import os
import pypandoc

for root, dirs, files in os.walk(r"C:\Users\valer\OneDrive\Desktop\LIT Hackathon\Emails Invoices and Bank Statements for Hackathon Problem Statement\Emails"):
    for file in files:
        if file.endswith(".docx"):
            docxFilename = (os.path.join(root, file))
       
            output = pypandoc.convert_file(docxFilename, 'plain', outputfile= str(docxFilename) + ".txt")
            assert output == ""
```
